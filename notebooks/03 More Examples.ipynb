```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# More plots types
## Scatter

Expects an X and Y set of same size.

Specialities to show:
- customizing markers
- set_aspect
- errorbars

#### Data

```python
X = np.random.normal(size=20)
dX = np.random.normal(0.3, 0.05, size=20)
Y = np.random.normal(size=20)
dY = np.random.uniform(0.5, 1.5, size=[20])*dX
```

#### Plot

```python
fig, ax = plt.subplots()
ax.errorbar(X, Y, xerr=dX, yerr=dY, ls='', lw=1, capsize=3, c='k', alpha=0.5)
ax.scatter(X, Y, s=30, marker=(5, 2, 45), label='2d data')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_xlim(-3, 3)
ax.set_ylim(-3, 3)
ax.set_aspect(1)
```

## Line-Plot
Expects at least a Y set, but can also make use of an X set.

Specialities to show:
- Line styles
- markers
- errorbars
- multiple scales?

#### Data

```python
df = 10
dphi = 0.05
X = np.linspace(0, 50, num=500)
Y = np.exp(-X/df) * np.sin(X)
Z = np.exp(-X/(1.5*df)) * np.cos((X-dphi*X))

sampling_rate = 100
y_smpl_x = X[::sampling_rate]
y_smpl_y = Y[::sampling_rate]
y_err = np.random.normal(0.3, 0.3, size=y_smpl_y.shape)
z_smpl_x = X[25::sampling_rate]
z_smpl_y = Z[25::sampling_rate]
z_err = np.random.normal(0.3, 0.3, size=z_smpl_y.shape)
```

#### Plot

```python
Y_lines = plt.plot(X, Y, '--', label='Y(X)')
Z_lines = plt.plot(X, Z, '-', label='Z(X)')
cy = Y_lines[0].get_color()
cz = Z_lines[0].get_color()

plt.errorbar(y_smpl_x, y_smpl_y, yerr=y_err, ls='', marker='o', color=cy, capsize=3, label='Y samples')
plt.errorbar(z_smpl_x, z_smpl_y, yerr=z_err, ls='', marker='o', color=cz, capsize=3, label='Z samples')
plt.xlabel('X')
plt.ylabel('Y(X), Z(X)')
plt.title('Dampening with dampening factor 10')
plt.legend(loc=4)
```

```python
plt.plot(Y, Z)
plt.errorbar(y_smpl_y, z_smpl_y, xerr=y_err, yerr=z_err, ls='', marker='o', capsize=3)
plt.gca().set_aspect(1)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xticks([-1,-0.5,0,0.5,1])
plt.yticks([-1,-0.5,0,0.5,1])
plt.grid()
plt.xlabel('Y(X)')
plt.ylabel('Z(X)')
plt.title('Phase-space')
```
