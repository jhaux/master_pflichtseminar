```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

#### Data

```python
df = 10
X = np.linspace(0, 50, num=500)
Y = np.exp(-X/df) * np.sin(X)
Z = np.exp(-X/df) * np.cos(X)
```

# Text and Annotations

Placing text in plots is easy with the Matplotlib. A lot of tools and helper functions let you put your text in your plot just as you want.

```python
fig, ax = plt.subplots(1,1)
ax.plot(X, Y)
ax.text(0.9, 0.9, 'Well, hello there!', 
        verticalalignment='top', horizontalalignment='right',
        transform=ax.transAxes)
```

$\Large{\LaTeX!!}$

Is totally possible and can be configured very detailed.

To not have to esquape all backslashes use raw strings via `r'somestring'`.

```python
equation = r'$f(x) = e^{-\frac{x}{\lambda}} \cdot \sin{x}$' \
         + '\n' \
         + r'$\lambda = {}$'.format(df)
fig, ax = plt.subplots(1,1)
ax.plot(X, Y)
ax.text(0.9, 0.9, equation, 
        verticalalignment='top', horizontalalignment='right',
        transform=ax.transAxes)
```

```python
import matplotlib as mpl
latex_custom_preamble = {
    "font.family": "serif", # use serif/main font for text elements
    "text.usetex": True,    # use inline math for ticks
    "text.latex.preamble": [
        r"\usepackage{amsmath}",  # for the align enivironment
        ]
}
mpl.rcParams.update(latex_custom_preamble)
```

```python
equation = r'\begin{align*}'
equation += r'f(x) &= e^{-\frac{x}{\lambda}} \cdot \sin{x} \\ '
equation += r'\lambda &= {} \text{{cm}} '.format(df)
equation += r'\end{align*}'

fig, ax = plt.subplots(1,1)
ax.plot(X, Y)
ax.text(0.9, 0.9, equation, 
        verticalalignment='top', horizontalalignment='right',
        transform=ax.transAxes,
        fontsize=15)
ax.set_xlabel('$x$ [cm]', fontsize=12)
ax.set_ylabel('$f(x)$', fontsize=12)
ax.set_title('\LaTeX', fontsize=20)
```

```python
equation = r'\begin{align*}'
equation += r'f(x) &= e^{-\frac{x}{\lambda}} \cdot \sin{x} \\ '
equation += r'\lambda &= {} \text{{cm}} '.format(df)
equation += r'\end{align*}'

datum = (X[300], Y[300])

fig, ax = plt.subplots(1,1)
ax.plot(X, Y)
ax.text(0.9, 0.9, equation, 
        verticalalignment='top', horizontalalignment='right',
        transform=ax.transAxes,
        fontsize=15)

ax.scatter(datum[0], datum[1], edgecolor='red')
ax.annotate('special point',
            xy=datum,
            xytext=(0.8, 0.1),
            textcoords=ax.transAxes,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle='arc3,rad=-0.3',
                            facecolor='black',
                            shrinkB=10))
ax.set_xlabel('$x$ [cm]', fontsize=12)
ax.set_ylabel('$f(x)$', fontsize=12)
ax.set_title('\LaTeX', fontsize=20)
```

```python
import matplotlib.patches as patches
```

```python
rect = patches.Rectangle((-0.5,-0.5), # Bottom left corner, 
                         1,           # width
                         1)           # height, all in data coordinates

fig, ax = plt.subplots(1,1)
ax.add_patch(rect)
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_aspect(1)
```

There are a lot of kwargs you can pass to the patch:
- hatch: some patter. e.g: '/', '\\', 'o', 'O', '+', '.', '*'
- fill: determine if to draw the fillcolor of the patch or not. True or False
- alpha: visibilty of the patch
- facecolor: color of the patch interior
- edgecolor: color of the border
- linewidth: thickness of the border
- and a lot more. See https://matplotlib.org/api/patches_api.html
